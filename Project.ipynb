Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

Preprocessing the test data set

```python
train_data_generator=ImageDataGenerator(rescale=1./255,shear_range=0.2,zoom_range=0.2,horizontal_flip=True)
training_test=train_data_generator.flow_from_directory("/content/Indian-monuments/images/train",target_size=(256,256),batch_size=32,class_mode="categorical")
```

Preprocessing the test data set

```python
test_data_generator=ImageDataGenerator(rescale=1./255)
testing_set=test_data_generator.flow_from_directory("/content/Indian-monuments/images/test",target_size=(256,256),batch_size=32,class_mode='categorical')
```

Buliding CNN initialization

```python
cnn=tf.keras.models.Sequential()
```

1sr layer

```python
cnn.add(tf.keras.layers.Conv2D(filters=256,kernel_size=3,activation="relu",input_shape=[256,256,3]))
```

pooling

```python
# max pooling
cnn.add(tf.keras.layers.MaxPool2D(strides=2,pool_size=2))
```

2nd convolutional layer

```python
cnn.add(tf.keras.layers.Conv2D(filters=256,kernel_size=3,activation="relu"))
cnn.add(tf.keras.layers.MaxPool2D(strides=2,pool_size=2))
```

Flattening

```python
cnn.add(tf.keras.layers.Flatten())
```

Full connectes layer

```python
cnn.add(tf.keras.layers.Dense(activation="relu",units=128))
```

Output layer

```python
cnn.add(tf.keras.layers.Dense(activation="softmax",units=24))
```

Compiling the cnn model

```python
cnn.compile(optimizer="adam", loss="categorical_crossentropy", metrics=['accuracy'])
```

training the model

```python
cnn.fit(x=training_test,validation_data=testing_set,epochs=100)
```

```python
training_test.class_indices
```

predicting a result

```python
from tensorflow.keras.preprocessing import image
test_img=image.load_img("/content/Indian-monuments/images/test/Ajanta Caves/4.jpg",target_size=(64,64))
test_img=image.img_to_array(test_img)
test_img=np.expand_dims(test_img,axis=0)
# making the right format
result=cnn.predict(test_img)
training_test.class_indices
print(result)
# if result[0][0]==1:
#     prediction="Dog"
# else:
#     prediction="Cat"
```
